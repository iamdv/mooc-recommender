```python
# Loading courses data from the CSV file

df_courses = pd.read_csv("../Data/main_coursera.csv")

df_cossim_01 = pd.read_csv("../Data/Model_CosSim_SingleVariable.csv")
```

```python
# Exploring data dimensionality, feature names, and feature types.

print(df_courses.shape,"\n")

print(df_courses.columns, "\n")

print(df_courses.info(), "\n")

print(df_courses.describe(), "\n")

print(df_cossim_01.describe(), "\n")
```

```python
# Sorting the cosine similarity results dataframe in descending order by the skill_score variable
df_cossim_01 = df_cossim_01.sort_values(['skill_score'], ascending = [False])

# Joining two dataframes - Courses and the Cosein Similarity Results based on the 'Course Id' variable. 
# Inner joins: Joins two tables with the common rows. This is a set operateion.
df_courses_score = df_courses.merge(df_cossim_01, on ='Course Id', how='inner')


df_courses_score = df_courses_score.sort_values(by = 'skill_score', ascending = False)

df_courses_score.sort_values(['skill_score', 'Rating'], ascending = [False, False] )

```
